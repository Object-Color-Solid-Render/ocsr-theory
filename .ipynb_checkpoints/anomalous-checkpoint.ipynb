```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# DeMarco, Pokorny & Smith (1992) versions of the Smith-Pokorny fundamentals
def loadcmfs(wavelengths, path):
    df = pd.read_csv(path) 
    
    ciergb = df.values 
    ciewavelengths = ciergb.T[0]
    ciergb = 10 **ciergb.T[1:] # un log scale
    # linearly interpolate to get ciergb in same shape as wavelengths
    ciergbs = np.zeros(shape=(3, len(wavelengths)))
    
    curridx = 0
    for i, wavelength in enumerate(wavelengths[:-2]):
        if ciewavelengths[curridx] + 5 < wavelength:
            curridx += 1
        if wavelength < ciewavelengths[0] or wavelength > ciewavelengths[-1]:
            ciergbs[:,i] = np.array([0,0,0])
        else:
            wavelength1 = ciewavelengths[curridx]
            rgb1 = ciergb[:,curridx]
            rgb2 = ciergb[:,curridx + 1]
            percent = (wavelength - wavelength1) / 5
            ciergbs[:,i] = rgb2 * percent + rgb1 * (1 - percent)
    return ciergbs
path = 'data/dpselog10.csv'
wavelengths = np.arange(400, 700, 1)
cmf = loadcmfs(wavelengths, path) # color matching functions now sampled at every 1 nm

lprimecone = np.roll(cmf[0], np.argmax(cmf[1]) - np.argmax(cmf[0]) + 10)
mprimecone = np.roll(cmf[1],  np.argmax(cmf[0]) - np.argmax(cmf[1]) - 6)

plt.title("cone fundamentals demarco pokorny and smith 1992")
plt.plot(wavelengths, cmf[0], color="red")
plt.plot(wavelengths, cmf[1] , color="green")
plt.plot(wavelengths, cmf[2], color="blue")
```

```python
protanomalous = [
    lprimecone,
    cmf[1],
    cmf[2]
] # L', M, S
deuteranomalous = [
    cmf[0],
    mprimecone,
    cmf[2]
] # L, M', S
```

```python
print("peak wavelengths of protanomaly", 
     wavelengths[np.argmax(lprimecone)],
    wavelengths[np.argmax(cmf[1])],
     wavelengths[np.argmax(cmf[2])],
     )
plt.title("cone fundamentals protanomaly")
plt.plot(wavelengths, protanomalous[0], color="magenta")
plt.plot(wavelengths, protanomalous[1] , color="green")
plt.plot(wavelengths, protanomalous[2], color="blue")
# plt.close()
```

```python
print("peak wavelengths of deuteranomaly",
      wavelengths[np.argmax(cmf[0])],
     wavelengths[np.argmax(mprimecone)],
     wavelengths[np.argmax(cmf[2])],
     )
plt.title("cone fundamentals deuteranomaly")
plt.plot(wavelengths, deuteranomalous[0], color="red")
plt.plot(wavelengths, deuteranomalous[1] , color="chartreuse")
plt.plot(wavelengths, deuteranomalous[2], color="blue")
```
